```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# make a time vector w/ 256 points
# and a source signal
N_cycles = 10.0
N_points = 512.0

t = np.arange(0,N_cycles*np.pi,np.pi*N_cycles/N_points)
signal = np.sin(t)

use_rect = False
if use_rect:
    threshold = -0.75
    signal[np.where(signal>=threshold)]=1.0
    signal[np.where(signal<threshold)]=-1.0
```

```python
# normalize the signal (not technically
# necessary for this example, but required
# for measuring correlation of physically
# different signals)
signal = signal/signal.std()

# generate two samples of the signal
# with a temporal offset:
N = 128
offset = 14
sample_1 = signal[:N]
sample_2 = signal[offset:N+offset]
```

```python
# determine the offset through cross-
# correlation
xc_num = np.abs(np.fft.ifft(np.fft.fft(sample_1)*np.fft.fft(sample_2).conjugate()))
offset_estimate = np.argmax(xc_num)
if offset_estimate>N//2:
    offset_estimate = offset_estimate - N
```

```python
offset_estimate
```

```python
fig, ax = plt.subplots()
ax.plot(sample_1, label='original')
ax.plot(sample_2, label='lagged')
plt.legend()
plt.show()
```

```python
f = 100 # sampling frequency
t = np.linspace(0, 1, f)
y = np.sin(15*t)
y_delayed = np.sin(15*t+np.pi/4)
y = y/y.std()
y_delayed = y_delayed/y_delayed.std()
```

```python
xc_num = np.abs(np.fft.ifft(np.fft.fft(y)*np.fft.fft(y_delayed).conjugate()))
N = len(t)
offset_estimate = np.argmax(xc_num)
if offset_estimate>N//2:
    offset_estimate = offset_estimate - N
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(t, y, label='original')
ax[0].plot(t, y_delayed, label='lagged')
ax[0].legend()

ax[1].plot(t, xc_num)

ax[1].vlines(offset_estimate*t.max()/f, ymin=0, ymax=xc_num.max())
plt.show()
```
