```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd
#plt.rcParams['figure.figsize'] = [16, 12]
plt.rcParams.update({'font.size': 18})
```

```python
data_dir = '../data/processed_25_02_2020/'
fname = 'trial5'

B_DELAY = 5
F_LENGTH=1095 # totally have second

# upload trial data
[pac_df, tapped_idx, labels] = pickle.load(open( data_dir + fname +'_essentials.pkl', 'rb') )
```

```python
# visualize the data
fig, ax = plt.subplots(figsize=(16,4))
ax.plot(pac_df.t, pac_df.pac)
ax.plot(pac_df.t[tapped_idx], pac_df.pac[tapped_idx], 'rx')
plt.show()
```

```python
# Let's look at one sample example
def get_tap_signal(n):
    return pac_df.pac[-B_DELAY+tapped_idx[n]:tapped_idx[n]+F_LENGTH]
```

```python
sample_signal = get_tap_signal(5)
```
