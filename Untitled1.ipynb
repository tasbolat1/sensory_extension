```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
import torch
torch.cuda.is_available()
```

```python
%matplotlib notebook
```

```python
data_dir = 'data/preprocessed_24_02_2020/start_optitrack_rb.pkl'
```

```python
data = pickle.load(open(data_dir, 'rb'))
```

```python
col_names = ['t', 'p_x', 'p_y', 'p_z', 'q_w', 'q_x', 'q_y', 'q_z']

marker_sizes = 8
for i in range(marker_sizes):
    col_names.append( 'marker_x_' + str(i))
    col_names.append( 'marker_y_' + str(i))
    col_names.append( 'marker_z_' + str(i))
    
all_table_data = []
for row in data:
    for rb in row[1:]:
        if rb[0] == 1:
            table_data = []
            table_data.append(row[0])
            for i in range(7):
                table_data.append(rb[i+1])
            for marker in rb[8:]:
                for j in range(3):
                    table_data.append(marker[j])
            all_table_data.append(table_data)
tabel_df = pd.DataFrame(data=all_table_data, columns=col_names)
```

```python
tabel_df
```

```python
# table markers stat
tabel_df.filter(like='marker' ).mean()
```

```python
tabel_df.filter(like='marker').std()
```

```python
tabel_df.filter(like='p' ).mean(), tabel_df.filter(like='p' ).std()
```

```python
# table markers stat
mean_table_markers = np.array([
    tabel_df.filter(like='marker_x' ).mean().values,
    tabel_df.filter(like='marker_y' ).mean().values,
    tabel_df.filter(like='marker_z' ).mean().values
])
mean_table_pos = tabel_df.filter(like='p' ).mean()
```

```python
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
# mapping from optitrack to world: x,y,z - > x, z, -y
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mean_table_markers[0,:], mean_table_markers[1,:], mean_table_markers[2,:])
ax.scatter(mean_table_pos[0], mean_table_pos[2], -mean_table_pos[1], 'rx')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
