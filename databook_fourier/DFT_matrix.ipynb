```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

n = 256
w = np.exp(-1j * 2 * np.pi / n)
```

```python
J, K = np.meshgrid( np.arange(n), np.arange(n) )
DFT = np.power(w, J*K)
DFT = np.real(DFT)

plt.imshow(DFT)
```
