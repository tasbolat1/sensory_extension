```python
peaks, _ = find_peaks( pac[4130:22350], height=0, distance=250*22)
```

```python
plt.plot(np.take( t[4130:22350], peaks), np.take( pac[4130:22350], peaks), 'x', label='pac 2.2 kHz')
plt.plot(t[4130:22350], pac[4130:22350], label='pac 2.2 kHz')
plt.show()
```

```python
peaks_raw, _ = find_peaks( pac, height=0, distance=150*22)
```

```python
peaks = []
for p in peaks_raw:
    if pac[p] > 1000:
        peaks.append(p)
```

```python
len(peaks)
```

```python
plt.plot(np.take( t, peaks), np.take( pac, peaks), 'x', label='pac 2.2 kHz')
plt.plot(t, pac, label='pac 2.2 kHz')
plt.show()
```

```python
top_opt = 'top_cal1_optitrack.pkl'
bot_opt = 'bottom_cal1_optitrack.pkl'
```

```python
# get z axis
opt = 'tap1_optitrack.pkl'
opt_df = read_optitrack(data_dir+opt)
top_df = read_optitrack(data_dir+top_opt)
bot_df = read_optitrack(data_dir+bot_opt)

top_optitrack_mean = top_df.pz.mean()
top_optitrack_std = top_df.pz.std()

bottom_optitrack_mean = bot_df.pz.mean()
bottom_optitrack_std = bot_df.pz.std()
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
find_nearest( opt_df.time.values, t[peaks[0]] )
```

```python
heights = []
for peak in peaks:
    heights.append( top_optitrack_mean - opt_df[opt_df.time == find_nearest( opt_df.time.values, t[peak] )].pz.values[0] )
```

```python
# heights being ranged
heights = np.array(heights)*100
heights = np.array(heights).astype(int)
```

```python
import tsfel
```

```python
fs = 2.2e3
```

```python
fmc = tsfel.feature_extraction.features.fft_mean_coeff(np.array( pac[-pe+peaks[0]:fe+peaks[0]] ), fs, nfreq=2200)
```

```python
tsfel.feature_extraction.features.power_bandwidth(np.array( pac[-pe+peaks[0]:fe+peaks[0]] ), fs)
```

```python
### Feature extraction
features_name = ["max_frequency", "max_power_spectrum", "power_bandwidth", "fft_mean_coeff", "spectral_spread", "fundamental_frequency", 'location']
```

```python
features = []
for i, peak in enumerate( peaks ):
    signal = np.array( pac[-pe+peak:fe+peak] )
    max_frequency = tsfel.feature_extraction.max_frequency(signal, fs)
    max_power_spectrum = tsfel.feature_extraction.max_power_spectrum(signal, fs)
    power_bandwidth = tsfel.feature_extraction.power_bandwidth(signal, fs)
    fft_mean_coeff = tsfel.feature_extraction.fft_mean_coeff(signal, fs, nfreq=1100)
    spectral_spread = tsfel.feature_extraction.spectral_spread(signal, fs)
    fundamental_frequency = tsfel.feature_extraction.fundamental_frequency(signal, fs)
    features.append([max_frequency, max_frequency, power_bandwidth, fft_mean_coeff, spectral_spread, fundamental_frequency, heights[i]])
```

```python
features_df = pd.DataFrame(columns=features_name, data=features)
```
