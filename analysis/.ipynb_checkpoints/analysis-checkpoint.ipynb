```python
# convert bag file to the readible csv
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

```python
data_dir = '../tap_data/table2/'
save_dir = '../results_imgs/'
fname = 'br2'
```

```python
def read_biotac(fname):
    data = pickle.load(open(data_dir+fname+'.pkl', 'rb'))
    columns = ['t', 'tdc', 'tac', 'pdc', 'pac', 'el']
    df = pd.DataFrame(data=data, columns=columns)
    df_el = pd.DataFrame(df.el.to_list(), index=df.index).add_prefix('el_').astype(int)
    df_pac = pd.DataFrame(df.pac.to_list(), index=df.index).add_prefix('pac_').astype(int)
    df.t = df.t-df.t[0]
    return df.join(df_el).join(df_pac)
```

```python
def time_interp(t):
    time_list = []
    for i in range(len(t)-1):
        time_list += list(np.linspace(t[i], t[i+1], 23))[:22]
    return time_list
```

```python
df = read_biotac(fname)
```

```python
from scipy.signal import find_peaks
```

```python
peaks, _ = find_peaks( df.pdc, height=0, distance=150)
```

```python
fig, ax = plt.subplots(2, figsize=(15,8))
ax[0].plot(df.t, df.pdc, label='pdc 100 Hz')
ax[0].plot(df.t[peaks], df.pdc[peaks], 'x')

ax[0].legend()
pac = sum(df.pac.to_list(), ())
t=time_interp(df.t.values)
ax[1].plot(t, pac[:-22], label='pac 2.2 kHz')
ax[1].legend()
plt.savefig('case2.png')
plt.show()
```

```python
r = range(1500,2050)
```

```python
fig, ax = plt.subplots(2, figsize=(15,8))
ax[0].plot(df.t.iloc[r], df.pdc.iloc[r], label='pdc 100 Hz')
ax[0].legend()
pac = sum(df.pac.iloc[r].to_list(), ())
t=time_interp(df.t.iloc[r].values)
ax[1].plot(t, pac[:-22], label='pac 2.2 kHz')
ax[1].legend()
plt.savefig('case2.png')
plt.show()
```
